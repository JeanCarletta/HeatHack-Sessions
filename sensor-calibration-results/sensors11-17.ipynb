## Sensors 11-17 

### Relative humidity


```python
#import plotly.offline as py
#py.init_notebook_mode(connected=True)

df = pd.read_csv("Sensors11-17.csv")
## this is not the way to change the axis labels, but I can't figure out the right way.
df['minute'] = df['obs']-1
df['%RH'] = df['Lascar h']
#df["timestamp"] = pd.to_datetime(df['created_at'])

fig = px.line(df, x='minute', y='%RH', title="Testing the calibration of our sensors 11-17 - the Lascar logger is shown in red.",range_y=[0,100], range_x=[0,60],color_discrete_sequence=['red'],width=800, height=400)
for x in range(11,17):
   fig.add_scatter(x = df['minute'], y = df['h' + str(x)],name=str(x))

fig.update_traces(marker_color="blue")
# fig.add_scatter(x = df['obs'], y = df['Lascar h'], name = 'lascar')
fig.add_hline(y='20',line_dash='dash')
fig.add_hline(y='80',line_dash='dash')
fig.show()
```

### Temperature

```python


# Using plotly.express

import plotly.express as px
import pandas as pd
df = pd.read_csv("Sensors11-17.csv")
## this is not the way to change the axis labels, but I can't figure out the right way.
df['minute'] = df['obs']-1
df['degC'] = df['Lascar t']

fig = px.line(df, x='minute', y='degC', title="Testing the calibration of our sensors 11-17 - the Lascar logger is shown in red.",range_y=[20,24], range_x=[0,60],color_discrete_sequence=['red'],width=800, height=400)
for x in range(11,17):
   fig.add_scatter(x = df['minute'], y = df['t' + str(x)],showlegend=False,name=str(x))

fig.update_traces(marker_color="blue")
fig.show()
```
